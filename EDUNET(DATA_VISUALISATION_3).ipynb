```python
import matplotlib.pyplot as plt

#Sample data for energy consumption over 6 months (in Mich)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption= [1200, 1300, 1100, 1500, 1400, 1600]


#Create a Line plot
plt.plot(months, energy_consumption, marker='o', color='b', linestyle='--')

#Add titles and Labels
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Month')
plt.ylabel('Energy Consumption (MWh)')
plt.show()
```

```python
#pair plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#creating a dataframe for pair plot
data={
    "Solar":[1200,1500,1300],
    "Wind":[3400,3600,3200],
    "Hydropower":[2900,3100,2800],
    "Biomass":[2500,2700,2400]
}
df=pd.DataFrame(data)

#create a pair plot
sns.pairplot(df)
plt.show()

#relationship between different features in data is shown by pair plot
```

```python
#create a heatmap to show the correlation matrix
correlation_matrix=df.corr()

#plot heatmap
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidths=0.5)

#add title
plt.title("Correlation between energy Sources")
plt.show()

#positive correlation-->ranging from 0 to 1-->hydropower and biomass have highest positive correlation

#--->look for highest values closest to 1 or -1
```

```python
#box plot
#min value-->lower quartile-->median-->upper quartile-->max
#                25%            50%         75%
#outliers-->It  is the noise/the unwanted or irrelvant values in data

#create a box plot
sns.boxplot(data=df)

#add title
plt.title("Distribution of Energy Consumption by Source")
plt.show()
```

```python
#create a violin plot
sns.violinplot(data=df)

#add title
plt.title("Violin Plot of Energy Consumption Distributions")
plt.show()

#to check the distribution among the various category
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Sample data for energy consumption and carbon emissions
energy_values = [100,200,300,400,500]
carbon_emissions= [10,20,25,40,50]

#create a dataframe with energy consumption ad carbon emissions
df_reg = pd.DataFrame({'Energy Consumption': energy_values,
                      'Carbon Emissions': carbon_emissions})

#Create a regression plot
sns.regplot(x='Energy Consumption', y='Carbon Emissions', data=df_reg)

#Add title
plt.title('Regression Plot of Energy Consumption vs Carbon Emissions')
plt.show()
```

```python
#3D PLOTTING OF LINES AND POINTS
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

fig=plt.figure()

#syntax for 3-D projection
ax=plt.axes(projection='3d')

#defining all 3 axes
z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

#plotting
ax.plot3D(x,y,z,'green')
ax.set_title('3D Line Plot')
plt.show()
```
